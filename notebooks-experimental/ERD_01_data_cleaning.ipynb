### **FROM 6 TXTs to 1 CSV & SAVE**

````python
from pathlib import Path
import pandas as pd

RAW_DIR = Path("../data/raw")
OUTPUT_FILE = Path("../data/processed/CONCAT_ValeursFoncieres.csv")

FILES = [
    "ValeursFoncieres-2020-S2.txt",
    "ValeursFoncieres-2021.txt",
    "ValeursFoncieres-2022.txt",
    "ValeursFoncieres-2023.txt",
    "ValeursFoncieres-2024.txt",
    "ValeursFoncieres-2025-S1.txt",
]

SEP = "|"
ENCODINGS = ["utf-8", "latin-1", "cp1252"]
CHUNK_SIZE = 50_000


def detect_encoding(filepath: Path) -> str:
    for enc in ENCODINGS:
        try:
            pd.read_csv(filepath, sep=SEP, nrows=5, encoding=enc)
            return enc
        except Exception:
            continue
    raise RuntimeError(f"Cannot read file: {filepath}")


def concat_files_chunked() -> None:
    OUTPUT_FILE.parent.mkdir(parents=True, exist_ok=True)

    first_write = True
    total_rows = 0

    for filename in FILES:
        path = RAW_DIR / filename

        if not path.exists():
            raise FileNotFoundError(f"Missing file: {path}")

        encoding = detect_encoding(path)
        print(f"\nProcessing {filename} with encoding={encoding}")

        reader = pd.read_csv(
            path,
            sep=SEP,
            dtype=str,
            encoding=encoding,
            low_memory=True,
            chunksize=CHUNK_SIZE
        )

        for chunk_idx, chunk in enumerate(reader, start=1):
            chunk.to_csv(
                OUTPUT_FILE,
                mode="w" if first_write else "a",
                index=False,
                header=first_write,
                encoding="utf-8"
            )

            first_write = False
            total_rows += len(chunk)

            print(
                f"  chunk {chunk_idx} written: {len(chunk):,} rows "
                f"| total: {total_rows:,}"
            )

            del chunk

    print("\nDone")
    print(f"Saved to: {OUTPUT_FILE}")
    print(f"Total rows written: {total_rows:,}")


if __name__ == "__main__":
    concat_files_chunked()
````

### **DROP COLUMNS**

````python
from pathlib import Path
import pandas as pd

INPUT_FILE = Path("../data/processed/CONCAT_ValeursFoncieres.csv")
OUTPUT_FILE = Path("../data/processed/DROP_ValeursFoncieres.csv")

CHUNK_SIZE = 50_000

cols_to_drop = [
    "Identifiant de document",
    "Reference document",
    "1 Articles CGI",
    "2 Articles CGI",
    "3 Articles CGI",
    "4 Articles CGI",
    "5 Articles CGI",
    "Identifiant local",
]

def drop_columns_chunked(input_file: Path, output_file: Path, cols_to_drop: list[str], chunk_size: int = 50_000) -> None:
    output_file.parent.mkdir(parents=True, exist_ok=True)

    first_write = True
    total_rows = 0

    for chunk_idx, chunk in enumerate(
        pd.read_csv(input_file, chunksize=chunk_size, dtype=str),
        start=1
    ):
        existing_cols_to_drop = [col for col in cols_to_drop if col in chunk.columns]
        chunk = chunk.drop(columns=existing_cols_to_drop)

        chunk.to_csv(
            output_file,
            mode="w" if first_write else "a",
            index=False,
            header=first_write,
            encoding="utf-8"
        )

        first_write = False
        total_rows += len(chunk)

        print(f"Chunk {chunk_idx} written -> {len(chunk):,} rows | total: {total_rows:,}")

    print("\nDone! Dropped columns")
    print(f"Saved file: {output_file}")
    print(f"Total rows written: {total_rows:,}")

drop_columns_chunked(INPUT_FILE, OUTPUT_FILE, cols_to_drop, CHUNK_SIZE)
````

### **FIXED INVALID VALUES FOR GOOD**

````python
from pathlib import Path
import pandas as pd

INPUT_FILE = Path("../data/processed/STANDARD_ValeursFoncieres.csv")
OUTPUT_FILE = Path("../data/processed/INVALIDS_ValeursFoncieres.csv")

CHUNK_SIZE = 50_000

date_cols = [
    "transaction_date",
]

numeric_float_cols = [
    "property_value",
    "building_real_surface",
    "land_surface",
    "lot_1_surface",
    "lot_2_surface",
    "lot_3_surface",
    "lot_4_surface",
    "lot_5_surface",
]

numeric_int_cols = [
    "lots_count",
    "main_rooms_count",
]

categorical_cols = [
    "transaction_id",
    "transaction_type",
    "street_number",
    "btq_code",
    "street_type",
    "street_id",
    "street_name",
    "postal_code",
    "com_name",
    "dep_code",
    "com_code",
    "section_prefix",
    "section",
    "plot_number",
    "volume_number",
    "property_type_code",
    "property_type",
    "land_nature",
    "land_nature_special",
    "lot_1",
    "lot_2",
    "lot_3",
    "lot_4",
    "lot_5",
]

missing_markers = {
    "MISSING", "", " ", "NA", "N/A", "None", "null", "NULL", "nan", "NaN"
}


def standardize_nulls(chunk: pd.DataFrame) -> pd.DataFrame:
    for col in chunk.columns:
        if chunk[col].dtype == "object" or str(chunk[col].dtype) == "string":
            chunk[col] = chunk[col].astype("string").str.strip()
            chunk[col] = chunk[col].replace(missing_markers, pd.NA)
    return chunk


def clean_date_columns(chunk: pd.DataFrame, cols: list[str]) -> pd.DataFrame:
    for col in cols:
        if col in chunk.columns:
            chunk[col] = pd.to_datetime(
                chunk[col],
                format="%d/%m/%Y",
                errors="coerce"
            )
    return chunk


def clean_float_columns(chunk: pd.DataFrame, cols: list[str]) -> pd.DataFrame:
    for col in cols:
        if col in chunk.columns:
            chunk[col] = (
                chunk[col]
                .astype("string")
                .str.replace(",", ".", regex=False)
                .str.strip()
            )
            chunk[col] = pd.to_numeric(chunk[col], errors="coerce")
    return chunk


def clean_int_columns(chunk: pd.DataFrame, cols: list[str]) -> pd.DataFrame:
    for col in cols:
        if col in chunk.columns:
            chunk[col] = (
                chunk[col]
                .astype("string")
                .str.strip()
            )
            chunk[col] = pd.to_numeric(chunk[col], errors="coerce").astype("Int64")
    return chunk


def clean_categorical_columns(chunk: pd.DataFrame, cols: list[str]) -> pd.DataFrame:
    for col in cols:
        if col in chunk.columns:
            chunk[col] = chunk[col].astype("string").str.strip()
    return chunk


def apply_custom_invalid_rules(chunk: pd.DataFrame) -> pd.DataFrame:
    # transaction_type: translate and standardize
    if "transaction_type" in chunk.columns:
        chunk["transaction_type"] = chunk["transaction_type"].replace({
            "Vente": "Sale",
            "Vente en l'état futur d'achèvement": "Sale in future state of completion",
            "Vente terrain à bâtir": "Sale of unbuilt land",
            "Echange": "Exchange",
            "Adjudication": "Auction",
            "Expropriation": "Expropriation",
        })

    # property_type: translate actual DVF categories seen in your data
    if "property_type" in chunk.columns:
        chunk["property_type"] = chunk["property_type"].replace({
            "Maison": "House",
            "Appartement": "Apartment",
            "Dépendance": "Outbuilding",
            "Local industriel. commercial ou assimilé": "Industrial or commercial premises",
        })

    # btq_code: clean invalid symbols and standardize lowercase b
    if "btq_code" in chunk.columns:
        chunk["btq_code"] = chunk["btq_code"].replace(["-", ".", "*"], pd.NA)
        chunk["btq_code"] = chunk["btq_code"].replace({"b": "B"})

    # building_real_surface and main_rooms_count:
    # use 0 when missing at this invalid-handling stage
    if "building_real_surface" in chunk.columns:
        chunk["building_real_surface"] = chunk["building_real_surface"].fillna(0.0)

    if "main_rooms_count" in chunk.columns:
        chunk["main_rooms_count"] = chunk["main_rooms_count"].fillna(0).astype("Int64")

    # create surface_type
    if {"building_real_surface", "land_surface"}.issubset(chunk.columns):
        b = chunk["building_real_surface"].fillna(0)
        l = chunk["land_surface"].fillna(0)

        chunk["surface_type"] = pd.NA
        chunk.loc[(b > 0) & (l == 0), "surface_type"] = "building"
        chunk.loc[(b == 0) & (l > 0), "surface_type"] = "land"
        chunk.loc[(b > 0) & (l > 0), "surface_type"] = "combined"

    return chunk


def clean_chunk(chunk: pd.DataFrame) -> pd.DataFrame:
    chunk = standardize_nulls(chunk)
    chunk = clean_date_columns(chunk, date_cols)
    chunk = clean_float_columns(chunk, numeric_float_cols)
    chunk = clean_int_columns(chunk, numeric_int_cols)
    chunk = clean_categorical_columns(chunk, categorical_cols)
    chunk = apply_custom_invalid_rules(chunk)
    return chunk


def clean_file_chunked(input_file: Path, output_file: Path, chunk_size: int = 50_000) -> None:
    output_file.parent.mkdir(parents=True, exist_ok=True)

    first_write = True
    total_rows = 0

    for chunk_idx, chunk in enumerate(
        pd.read_csv(input_file, chunksize=chunk_size, dtype=str, low_memory=False),
        start=1
    ):
        chunk = clean_chunk(chunk)

        chunk.to_csv(
            output_file,
            mode="w" if first_write else "a",
            index=False,
            header=first_write,
            encoding="utf-8"
        )

        first_write = False
        total_rows += len(chunk)

        print(f"Chunk {chunk_idx} written -> {len(chunk):,} rows | total: {total_rows:,}")

    print("\nDone! Handled invalid values")
    print(f"Saved file: {output_file}")
    print(f"Total rows written: {total_rows:,}")


clean_file_chunked(INPUT_FILE, OUTPUT_FILE, CHUNK_SIZE)
````

### **DEAL with INVALID VALUES**

````python
from pathlib import Path
import pandas as pd

INPUT_FILE = Path("../data/processed/STANDARD_ValeursFoncieres.csv")
OUTPUT_FILE = Path("../data/processed/INVALIDS_ValeursFoncieres.csv")

CHUNK_SIZE = 50_000

date_cols = [
    "transaction_date",
]

numeric_float_cols = [
    "property_value",
    "building_real_surface",
    "land_surface",
    "lot_1_surface",
    "lot_2_surface",
    "lot_3_surface",
    "lot_4_surface",
    "lot_5_surface",
]

numeric_int_cols = [
    "lots_count",
    "main_rooms_count",
]

categorical_cols = [
    "transaction_id",
    "transaction_type",
    "street_number",
    "btq_code",
    "street_type",
    "street_id",
    "street_name",
    "postal_code",
    "com_name",
    "dep_code",
    "com_code",
    "section_prefix",
    "section",
    "plot_number",
    "volume_number",
    "property_type_code",
    "property_type",
    "land_nature",
    "land_nature_special",
    "lot_1",
    "lot_2",
    "lot_3",
    "lot_4",
    "lot_5",
]

missing_markers = {
    "MISSING", "", " ", "NA", "N/A", "None", "null", "NULL", "nan", "NaN"
}


def clean_string_columns(chunk: pd.DataFrame) -> pd.DataFrame:
    object_cols = chunk.select_dtypes(include="object").columns

    for col in object_cols:
        chunk[col] = chunk[col].astype(str).str.strip()
        chunk[col] = chunk[col].replace(list(missing_markers), pd.NA)

    return chunk


def clean_date_columns(chunk: pd.DataFrame, cols: list[str]) -> pd.DataFrame:
    for col in cols:
        if col in chunk.columns:
            chunk[col] = pd.to_datetime(
                chunk[col],
                format="%d/%m/%Y",
                errors="coerce"
            )
    return chunk


def clean_float_columns(chunk: pd.DataFrame, cols: list[str]) -> pd.DataFrame:
    for col in cols:
        if col in chunk.columns:
            chunk[col] = (
                chunk[col]
                .astype("string")
                .str.replace(",", ".", regex=False)
            )
            chunk[col] = pd.to_numeric(chunk[col], errors="coerce")
    return chunk


def clean_int_columns(chunk: pd.DataFrame, cols: list[str]) -> pd.DataFrame:
    for col in cols:
        if col in chunk.columns:
            chunk[col] = pd.to_numeric(chunk[col], errors="coerce").astype("Int64")
    return chunk


def clean_categorical_columns(chunk: pd.DataFrame, cols: list[str]) -> pd.DataFrame:
    for col in cols:
        if col in chunk.columns:
            chunk[col] = chunk[col].astype("string").str.strip()
            chunk[col] = chunk[col].replace(list(missing_markers), pd.NA)
    return chunk


def apply_custom_invalid_rules(chunk: pd.DataFrame) -> pd.DataFrame:
    # ---------------------------------------------------
    # transaction_type: translate and standardize
    # ---------------------------------------------------
    if "transaction_type" in chunk.columns:
        chunk["transaction_type"] = chunk["transaction_type"].replace({
            "Vente": "Sale",
            "Vente en l'état futur d'achèvement": "Sale in future state of completion",
            "Vente terrain à bâtir": "Sale of unbuilt land",
            "Echange": "Exchange",
            "Adjudication": "Auction",
            "Expropriation": "Expropriation",
        })

    # ---------------------------------------------------
    # property_type: translate and standardize
    # ---------------------------------------------------
    if "property_type" in chunk.columns:
        chunk["property_type"] = chunk["property_type"].replace({
            "Maison": "House",
            "Appartement": "Apartment",
            "Terrain": "Land",
            "Local commercial": "Commercial",
            "Bureau": "Office",
            "Autre": "Other",
        })
    # ---------------------------------------------------
    # surface_type: translate and standardize
    # ---------------------------------------------------
    # surface_type = "building" if building_real_surface > 0 and land_surface is 0
    # surface_type = "land" if building_real_surface is 0 and land_surface > 0
    # surface_type = "combined" if building_real_surface > 0 and land_surface > 0

    # ---------------------------------------------------
    # btq_code: clean invalid codes and standardize case
    # ---------------------------------------------------
    if "btq_code" in chunk.columns:
        invalid_values = ["-", ".", "*"]
        chunk["btq_code"] = chunk["btq_code"].replace(invalid_values, pd.NA)
        chunk["btq_code"] = chunk["btq_code"].replace({"b": "B"})

    # ---------------------------------------------------
    # building_real_surface: fill missing with 0
    # ---------------------------------------------------
    if "building_real_surface" in chunk.columns:
        chunk["building_real_surface"] = chunk["building_real_surface"].fillna(0)

    # ---------------------------------------------------
    # main_rooms_count: fill missing with 0
    # ---------------------------------------------------
    if "main_rooms_count" in chunk.columns:
        chunk["main_rooms_count"] = chunk["main_rooms_count"].fillna(0)

    return chunk


def clean_chunk(chunk: pd.DataFrame) -> pd.DataFrame:
    chunk = clean_string_columns(chunk)
    chunk = clean_date_columns(chunk, date_cols)
    chunk = clean_float_columns(chunk, numeric_float_cols)
    chunk = clean_int_columns(chunk, numeric_int_cols)
    chunk = clean_categorical_columns(chunk, categorical_cols)
    chunk = apply_custom_invalid_rules(chunk)
    return chunk


def clean_file_chunked(input_file: Path, output_file: Path, chunk_size: int = 50_000) -> None:
    output_file.parent.mkdir(parents=True, exist_ok=True)

    first_write = True
    total_rows = 0

    for chunk_idx, chunk in enumerate(
        pd.read_csv(input_file, chunksize=chunk_size, dtype=str),
        start=1
    ):
        chunk = clean_chunk(chunk)

        chunk.to_csv(
            output_file,
            mode="w" if first_write else "a",
            index=False,
            header=first_write,
            encoding="utf-8"
        )

        first_write = False
        total_rows += len(chunk)

        print(f"Chunk {chunk_idx} written -> {len(chunk):,} rows | total: {total_rows:,}")

    print("\nDone! Handled invalid values")
    print(f"Saved file: {output_file}")
    print(f"Total rows written: {total_rows:,}")


clean_file_chunked(INPUT_FILE, OUTPUT_FILE, CHUNK_SIZE)
````

### **NULL COUNT vs UNIQUE VALUES**

````python
from collections import defaultdict
import pandas as pd

FILE = "../data/processed/INVALIDS_ValeursFoncieres.csv"

unique_values = defaultdict(set)
null_counts = defaultdict(int)
total_rows = 0

for chunk in pd.read_csv(
    FILE,
    chunksize=50_000,
    dtype=str,          # ✅ avoids mixed types warning
    low_memory=False
):
    total_rows += len(chunk)

    for col in chunk.columns:
        # ✅ Count nulls (DO NOT drop them)
        null_counts[col] += chunk[col].isna().sum()

        # ✅ Collect only non-null values (nulls are still preserved in dataset)
        unique_values[col].update(
            chunk.loc[chunk[col].notna(), col].unique()
        )

# --------------------------------------------------
# Display results
# --------------------------------------------------
print(f"\nTotal rows processed: {total_rows:,}")
print(100 * "-")

for col in unique_values:
    values = list(unique_values[col])
    null_pct = (null_counts[col] / total_rows) * 100

    print(f"\n➡️ {col}:")
    print(f"Unique values count: {len(values)}")
    print(f"Null count: {null_counts[col]} ({null_pct:.2f}%)")
    print(f"Sample values: {values[:120]}")

    print(100 * "-")
````

### **CONVERT DATA TYPES**

````python
from pathlib import Path
import pandas as pd

INPUT_FILE = Path("../data/processed/INVALIDS_ValeursFoncieres.csv")
OUTPUT_FILE = Path("../data/processed/TYPES_ValeursFoncieres.csv")

CHUNK_SIZE = 50_000

# Columns by target type
string_cols = [
    "transaction_id",
    "transaction_type",
    "street_number",
    "btq_code",
    "street_type",
    "street_id",
    "street_name",
    "postal_code",
    "com_name",
    "dep_code",
    "com_code",
    "section_prefix",
    "section",
    "plot_number",
    "volume_number",
    "property_type_code",
    "property_type",
    "land_nature",
    "land_nature_special",
    "surface_type",
    "lot_1",
    "lot_2",
    "lot_3",
    "lot_4",
    "lot_5"
]

date_cols = [
    "transaction_date",
]

float_cols = [
    "property_value",
    "building_real_surface",
    "land_surface",
    "lot_1_surface",
    "lot_2_surface",
    "lot_3_surface",
    "lot_4_surface",
    "lot_5_surface"
]

int_cols = [
    "lots_count",
    "main_rooms_count",
]


def convert_chunk_types(chunk: pd.DataFrame) -> pd.DataFrame:
    # String / categorical columns
    for col in string_cols:
        if col in chunk.columns:
            chunk[col] = chunk[col].astype("string").str.strip()

    # Date columns
    for col in date_cols:
        if col in chunk.columns:
            chunk[col] = pd.to_datetime(
                chunk[col],
                errors="coerce"
            )

    # Float columns
    for col in float_cols:
        if col in chunk.columns:
            if chunk[col].dtype == "object" or str(chunk[col].dtype) == "string":
                chunk[col] = (
                    chunk[col]
                    .astype("string")
                    .str.replace(",", ".", regex=False)
                    .str.strip()
                )
            chunk[col] = pd.to_numeric(chunk[col], errors="coerce").astype("float64")

    # Integer columns
    for col in int_cols:
        if col in chunk.columns:
            if chunk[col].dtype == "object" or str(chunk[col].dtype) == "string":
                chunk[col] = chunk[col].astype("string").str.strip()
            chunk[col] = pd.to_numeric(chunk[col], errors="coerce").astype("Int64")

    return chunk


def convert_file_types_chunked(
    input_file: Path,
    output_file: Path,
    chunk_size: int = 50_000
) -> None:
    output_file.parent.mkdir(parents=True, exist_ok=True)

    first_write = True
    total_rows = 0

    for chunk_idx, chunk in enumerate(
        pd.read_csv(input_file, chunksize=chunk_size, dtype=str),
        start=1
    ):
        chunk = convert_chunk_types(chunk)

        chunk.to_csv(
            output_file,
            mode="w" if first_write else "a",
            index=False,
            header=first_write,
            encoding="utf-8"
        )

        first_write = False
        total_rows += len(chunk)

        print(f"Chunk {chunk_idx} written -> {len(chunk):,} rows | total: {total_rows:,}")

    print("\nDone! Converted column data types")
    print(f"Saved file: {output_file}")
    print(f"Total rows written: {total_rows:,}")


convert_file_types_chunked(INPUT_FILE, OUTPUT_FILE, CHUNK_SIZE)
````

````python
from pathlib import Path
import pandas as pd

INPUT_FILE = Path("../data/processed/TYPES_ValeursFoncieres.csv")
OUTPUT_FILE = Path("../data/processed/NULLS_ValeursFoncieres.csv")

CHUNK_SIZE = 50_000

dtype_map = {
    "transaction_id": "string",
    "transaction_date": "string",
    "transaction_type": "string",
    "property_value": "float64",
    "street_number": "string",
    "btq_code": "string",
    "street_type": "string",
    "street_id": "string",
    "street_name": "string",
    "postal_code": "string",
    "com_name": "string",
    "dep_code": "string",
    "com_code": "string",
    "section_prefix": "string",
    "section": "string",
    "plot_number": "string",
    "volume_number": "string",
    "lot_1": "string",
    "lot_1_surface": "float64",
    "lot_2": "string",
    "lot_2_surface": "float64",
    "lot_3": "string",
    "lot_3_surface": "float64",
    "lot_4": "string",
    "lot_4_surface": "float64",
    "lot_5": "string",
    "lot_5_surface": "float64",
    "lots_count": "Int64",
    "property_type_code": "string",
    "property_type": "string",
    "building_real_surface": "float64",
    "main_rooms_count": "Int64",
    "land_nature": "string",
    "land_nature_special": "string",
    "land_surface": "float64",
    "surface_type": "string",
}

categorical_cols = [
    "transaction_id",
    "transaction_date",
    "transaction_type",
    "street_number",
    "btq_code",
    "street_type",
    "street_id",
    "street_name",
    "postal_code",
    "com_name",
    "dep_code",
    "com_code",
    "section_prefix",
    "section",
    "plot_number",
    "volume_number",
    "lot_1",
    "lot_2",
    "lot_3",
    "lot_4",
    "lot_5",
    "property_type_code",
    "property_type",
    "land_nature",
    "land_nature_special",
    "surface_type",
]

numeric_float_cols = [
    "property_value",
    "lot_1_surface",
    "lot_2_surface",
    "lot_3_surface",
    "lot_4_surface",
    "lot_5_surface",
    "building_real_surface",
    "land_surface",
]

numeric_int_cols = [
    "lots_count",
    "main_rooms_count",
]


def handle_nulls(chunk: pd.DataFrame) -> pd.DataFrame:
    # target cannot be imputed safely
    chunk = chunk.dropna(subset=["property_value"]).copy()

    # numeric: smarter fills first
    chunk.loc[:, "building_real_surface"] = chunk["building_real_surface"].fillna(0.0)

    for col in ["lot_1_surface", "lot_2_surface", "lot_3_surface", "lot_4_surface", "lot_5_surface"]:
        if col in chunk.columns:
            chunk.loc[:, col] = chunk[col].fillna(0.0)

    # main_rooms_count:
    # if missing and building_real_surface exists =>0, set main_rooms_count = 0
    mask_rooms = (
        chunk["main_rooms_count"].isna() &
        (chunk["building_real_surface"] >= 0)
    )
    chunk.loc[mask_rooms, "main_rooms_count"] = 0
    


    # land_surface:
    # if missing and building_real_surface exists, set to 0
    mask_land = (
        chunk["land_surface"].isna() &
        (chunk["building_real_surface"] > 0)
    )
    chunk.loc[mask_land, "land_surface"] = 0.0

    # fallback numeric fills
    for col in numeric_float_cols:
        if col in chunk.columns and col != "property_value":
            chunk.loc[:, col] = chunk[col].fillna(0.0)

    for col in numeric_int_cols:
        if col in chunk.columns:
            chunk.loc[:, col] = chunk[col].fillna(0)

    # categorical fills
    for col in categorical_cols:
        if col in chunk.columns:
            chunk.loc[:, col] = chunk[col].fillna("Unknown")

    # restore numeric dtypes
    for col in numeric_float_cols:
        if col in chunk.columns:
            chunk.loc[:, col] = pd.to_numeric(chunk[col], errors="coerce")

    for col in numeric_int_cols:
        if col in chunk.columns:
            chunk.loc[:, col] = pd.to_numeric(chunk[col], errors="coerce").astype("Int64")

    return chunk


def process_nulls_chunked(input_file: Path, output_file: Path, chunk_size: int = 50_000) -> None:
    output_file.parent.mkdir(parents=True, exist_ok=True)

    first_write = True
    total_rows_read = 0
    total_rows_written = 0
    total_rows_dropped = 0

    for i, chunk in enumerate(
        pd.read_csv(
            input_file,
            chunksize=chunk_size,
            dtype=dtype_map,
            low_memory=False
        ),
        start=1
    ):
        rows_before = len(chunk)
        total_rows_read += rows_before

        chunk = handle_nulls(chunk)

        rows_after = len(chunk)
        total_rows_written += rows_after
        total_rows_dropped += (rows_before - rows_after)

        chunk.to_csv(
            output_file,
            mode="w" if first_write else "a",
            index=False,
            header=first_write,
            encoding="utf-8"
        )

        first_write = False

        print(
            f"Chunk {i} processed -> "
            f"read: {rows_before:,} | kept: {rows_after:,} | "
            f"dropped: {rows_before - rows_after:,} | total kept: {total_rows_written:,}"
        )

    print("\nDone! Null values handled")
    print(f"Saved file: {output_file}")
    print(f"Total rows read: {total_rows_read:,}")
    print(f"Total rows dropped: {total_rows_dropped:,}")
    print(f"Total rows written: {total_rows_written:,}")


process_nulls_chunked(INPUT_FILE, OUTPUT_FILE, CHUNK_SIZE)
````

### **DUPLICATE ROWS**

````python
from pathlib import Path
import pandas as pd
import hashlib

FILE = Path("../data/processed/NULLS_ValeursFoncieres.csv")
CHUNK_SIZE = 50_000

dtype_map = {
    # categorical columns
    "transaction_id": "string",
    "transaction_date": "string",
    "transaction_type": "string",
    "street_number": "string",
    "btq_code": "string",
    "street_type": "string",
    "street_id": "string",
    "street_name": "string",
    "postal_code": "string",
    "com_name": "string",
    "dep_code": "string",
    "com_code": "string",
    "section_prefix": "string",
    "section": "string",
    "plot_number": "string",
    "volume_number": "string",
    "property_type_code": "string",
    "property_type": "string",
    "land_nature": "string",
    "land_nature_special": "string",
    "surface_type": "string",
    # numeric columns
    "property_value": "float64",
    "building_real_surface": "float64",
    "land_surface": "float64",
    "lots_count": "Int64",
    "main_rooms_count": "Int64",
}

def row_hash(row: pd.Series) -> str:
    values = row.fillna("<NA>").astype(str).tolist()
    row_str = "||".join(values)
    return hashlib.md5(row_str.encode("utf-8")).hexdigest()

total_rows = 0
duplicate_rows_count = 0
seen_hashes = set()

null_counts = None
columns = None
duplicate_samples = []

for chunk in pd.read_csv(
    FILE,
    chunksize=CHUNK_SIZE,
    dtype=dtype_map,
    low_memory=False
):
    if columns is None:
        columns = chunk.columns.tolist()
        null_counts = {col: 0 for col in columns}

    total_rows += len(chunk)

    # Null counts
    for col in columns:
        null_counts[col] += chunk[col].isna().sum()

    # Duplicate detection across all chunks
    chunk_hashes = chunk.apply(row_hash, axis=1)

    is_duplicate = chunk_hashes.isin(seen_hashes)
    duplicate_rows_count += is_duplicate.sum()

    # Save a few duplicate row samples for display
    if is_duplicate.any() and len(duplicate_samples) < 20:
        sample_dups = chunk.loc[is_duplicate].head(20 - len(duplicate_samples))
        duplicate_samples.append(sample_dups)

    # Add only new hashes
    new_hashes = chunk_hashes.loc[~is_duplicate]
    seen_hashes.update(new_hashes.tolist())

# Build null percentage summary
null_pct = {
    col: round(null_counts[col] / total_rows * 100, 2)
    for col in columns
}

nulls_df = pd.DataFrame({
    "column": columns,
    "null_count": [null_counts[col] for col in columns],
    "null_pct": [null_pct[col] for col in columns]
}).sort_values("null_pct", ascending=False)

# Print results
print(f"Number of duplicate rows: {duplicate_rows_count}")
print(100 * "-")
print("Nulls percentage in each column:")
display(nulls_df)
print(100 * "-")
print(f"Duplicate rows percentage: {duplicate_rows_count / total_rows * 100:.2f}%")
print(100 * "-")

# Display duplicate samples if any
if duplicate_rows_count > 0:
    print("\nDuplicate row samples:")
    display(pd.concat(duplicate_samples, ignore_index=True))
else:
    print("\nNo duplicate rows found.")
````

### **DEAL with DUPLICATE ROWS**

````python
from pathlib import Path
import pandas as pd
import hashlib

INPUT_FILE = Path("../data/processed/NULLS_ValeursFoncieres.csv")
OUTPUT_FILE = Path("../data/processed/DUPLICATES_ValeursFoncieres.csv")

CHUNK_SIZE = 50_000

dtype_map = {
    "transaction_id": "string",
    "transaction_date": "string",
    "transaction_type": "string",
    "property_value": "float64",
    "street_number": "string",
    "btq_code": "string",
    "street_type": "string",
    "street_id": "string",
    "street_name": "string",
    "postal_code": "string",
    "com_name": "string",
    "dep_code": "string",
    "com_code": "string",
    "section_prefix": "string",
    "section": "string",
    "plot_number": "string",
    "volume_number": "string",
    "lot_1": "string",
    "lot_1_surface": "float64",
    "lot_2": "string",
    "lot_2_surface": "float64",
    "lot_3": "string",
    "lot_3_surface": "float64",
    "lot_4": "string",
    "lot_4_surface": "float64",
    "lot_5": "string",
    "lot_5_surface": "float64",
    "lots_count": "Int64",
    "property_type_code": "string",
    "property_type": "string",
    "building_real_surface": "float64",
    "main_rooms_count": "Int64",
    "land_nature": "string",
    "land_nature_special": "string",
    "land_surface": "float64",
    "surface_type": "string",
}

def row_hash(row: pd.Series) -> str:
    values = row.fillna("<NA>").astype(str).tolist()
    row_str = "||".join(values)
    return hashlib.md5(row_str.encode("utf-8")).hexdigest()

def remove_true_duplicates_chunked(input_file: Path, output_file: Path, chunk_size: int = 50_000) -> None:
    output_file.parent.mkdir(parents=True, exist_ok=True)

    seen_hashes = set()
    first_write = True

    total_rows_read = 0
    total_rows_written = 0
    total_duplicates_removed = 0

    for chunk_idx, chunk in enumerate(
        pd.read_csv(
            input_file,
            chunksize=chunk_size,
            dtype=dtype_map,
            low_memory=False
        ),
        start=1
    ):
        rows_before = len(chunk)
        total_rows_read += rows_before

        # Remove exact duplicates inside the chunk
        chunk = chunk.drop_duplicates().copy()

        # Hash full rows
        hashes = chunk.apply(row_hash, axis=1)

        # Keep only rows not seen before
        keep_mask = ~hashes.isin(seen_hashes)
        chunk_kept = chunk.loc[keep_mask].copy()

        # Update seen hashes
        seen_hashes.update(hashes.loc[keep_mask].tolist())

        rows_after = len(chunk_kept)
        removed_here = rows_before - rows_after

        total_rows_written += rows_after
        total_duplicates_removed += removed_here

        chunk_kept.to_csv(
            output_file,
            mode="w" if first_write else "a",
            index=False,
            header=first_write,
            encoding="utf-8"
        )

        first_write = False

        print(
            f"Chunk {chunk_idx} processed -> "
            f"read: {rows_before:,} | kept: {rows_after:,} | "
            f"removed: {removed_here:,} | total kept: {total_rows_written:,}"
        )

    print("\nDone! Removed true duplicates only")
    print(f"Saved file: {output_file}")
    print(f"Total rows read: {total_rows_read:,}")
    print(f"Total duplicates removed: {total_duplicates_removed:,}")
    print(f"Total rows written: {total_rows_written:,}")

remove_true_duplicates_chunked(INPUT_FILE, OUTPUT_FILE, CHUNK_SIZE)
````

````python
from pathlib import Path
import pandas as pd

INPUT_FILE = Path("../data/processed/NULLS_ValeursFoncieres.csv")
OUTPUT_FILE = Path("../data/processed/DUPLICATES_ValeursFoncieres.csv")

CHUNK_SIZE = 50_000

dtype_map = {
    "transaction_id": "string",
    "transaction_date": "string",
    "transaction_type": "string",
    "property_value": "float64",
    "street_number": "string",
    "btq_code": "string",
    "street_type": "string",
    "street_id": "string",
    "street_name": "string",
    "postal_code": "string",
    "com_name": "string",
    "dep_code": "string",
    "com_code": "string",
    "section_prefix": "string",
    "section": "string",
    "plot_number": "string",
    "volume_number": "string",
    "lot_1": "string",
    "lot_1_surface": "float64",
    "lot_2": "string",
    "lot_2_surface": "float64",
    "lot_3": "string",
    "lot_3_surface": "float64",
    "lot_4": "string",
    "lot_4_surface": "float64",
    "lot_5": "string",
    "lot_5_surface": "float64",
    "lots_count": "Int64",
    "property_type_code": "string",
    "property_type": "string",
    "building_real_surface": "float64",
    "main_rooms_count": "Int64",
    "land_nature": "string",
    "land_nature_special": "string",
    "land_surface": "float64",
    "surface_type": "string",
}

def remove_duplicates_chunked(input_file, output_file, chunk_size):
    output_file.parent.mkdir(parents=True, exist_ok=True)

    seen = set()
    first_write = True
    total_kept = 0

    for i, chunk in enumerate(
        pd.read_csv(
            input_file,
            chunksize=chunk_size,
            dtype=dtype_map,
            low_memory=False
        ),
        start=1
    ):
        # Convert rows to tuples (fast + consistent)
        rows = chunk.astype(str).agg(tuple, axis=1)

        # Keep first occurrence only
        mask = ~rows.isin(seen)
        chunk = chunk.loc[mask]

        # Update seen rows
        seen.update(rows[mask])

        total_kept += len(chunk)

        chunk.to_csv(
            output_file,
            mode="w" if first_write else "a",
            index=False,
            header=first_write,
            encoding="utf-8"
        )

        first_write = False

        print(f"Chunk {i} -> kept: {len(chunk):,} | total: {total_kept:,}")

    print("\nDone! Removed true duplicate rows")
    print(f"Saved file: {output_file}")
    print(f"Total rows written: {total_kept:,}")

remove_duplicates_chunked(INPUT_FILE, OUTPUT_FILE, CHUNK_SIZE)
````

# MACHINE LEARNING

````python
from pathlib import Path
import pandas as pd

INPUT_FILE = Path("../data/processed/CLEAN_ValeursFoncieres.csv")
OUTPUT_FILE = Path("../data/processed/ML_ValeursFoncieres.csv")

CHUNK_SIZE = 50_000


def preprocess_ml(chunk: pd.DataFrame) -> pd.DataFrame:
    # ---------------------------------------------------
    # Drop columns that leak target or are not useful
    # ---------------------------------------------------
    cols_to_drop = [
        "value_per_m2",      # TARGET LEAKAGE: property_value / property_surface
        "transaction_date",  # not needed for your app goal
        "com_name",          # free-text location label, redundant with codes
    ]

    chunk = chunk.drop(columns=[c for c in cols_to_drop if c in chunk.columns])

    # ---------------------------------------------------
    # Keep only rows where target exists
    # ---------------------------------------------------
    chunk = chunk.dropna(subset=["property_value"]).copy()

    return chunk


def process_ml_chunked(input_file: Path, output_file: Path, chunk_size: int = 50_000) -> None:
    output_file.parent.mkdir(parents=True, exist_ok=True)

    first_write = True
    total_rows = 0

    for i, chunk in enumerate(
        pd.read_csv(input_file, chunksize=chunk_size, low_memory=False),
        start=1
    ):
        chunk = preprocess_ml(chunk)

        chunk.to_csv(
            output_file,
            mode="w" if first_write else "a",
            index=False,
            header=first_write,
            encoding="utf-8"
        )

        first_write = False
        total_rows += len(chunk)

        print(f"Chunk {i} processed -> {len(chunk):,} rows | total: {total_rows:,}")

    print("\nDone! ML dataset created")
    print(f"Saved file: {output_file}")
    print(f"Total rows written: {total_rows:,}")


process_ml_chunked(INPUT_FILE, OUTPUT_FILE, CHUNK_SIZE)
````

### CREATE ANALYSIS_ValeursFoncieres.csv

````python
from pathlib import Path
import pandas as pd
import numpy as np

INPUT_FILE = Path("../data/processed/CLEAN_ValeursFoncieres.csv")
OUTPUT_FILE = Path("../data/processed/ANALYSIS_ValeursFoncieres.csv")

CHUNK_SIZE = 50_000


def preprocess_analysis(chunk: pd.DataFrame) -> pd.DataFrame:

    # ---------------------------------------------------
    # DROP USELESS COLUMNS
    # ---------------------------------------------------
    cols_to_drop = [
        "transaction_id",
        "street_name",
        "street_id",
        "street_number",
        "btq_code",
        "plot_number",
        "volume_number",
        "section",
        "section_prefix",
        "lot_1", "lot_2", "lot_3", "lot_4", "lot_5",
        "lot_1_surface", "lot_2_surface", "lot_3_surface",
        "lot_4_surface", "lot_5_surface",
    ]

    chunk = chunk.drop(columns=[c for c in cols_to_drop if c in chunk.columns])

    # ---------------------------------------------------
    # FEATURE ENGINEERING (for analysis)
    # ---------------------------------------------------

    # Property surface
    chunk["property_surface"] = (
        chunk["building_real_surface"].fillna(0) +
        chunk["land_surface"].fillna(0)
    )

    # €/m²
    chunk["value_per_m2"] = chunk["property_value"] / chunk["property_surface"]

    # Avoid inf values
    chunk["value_per_m2"] = chunk["value_per_m2"].replace([np.inf, -np.inf], np.nan)

    # INSEE code (VERY IMPORTANT)
    chunk["insee_code"] = chunk["dep_code"] + chunk["com_code"]

    return chunk


def process_analysis_chunked(input_file, output_file, chunk_size):

    output_file.parent.mkdir(parents=True, exist_ok=True)

    first_write = True
    total_rows = 0

    for i, chunk in enumerate(
        pd.read_csv(input_file, chunksize=chunk_size),
        start=1
    ):
        chunk = preprocess_analysis(chunk)

        chunk.to_csv(
            output_file,
            mode="w" if first_write else "a",
            index=False,
            header=first_write,
            encoding="utf-8"
        )

        first_write = False
        total_rows += len(chunk)

        print(f"Chunk {i} processed -> {len(chunk):,} rows | total: {total_rows:,}")

    print("\nDone! ANALYSIS dataset created")
    print(f"Saved file: {output_file}")
    print(f"Total rows: {total_rows:,}")


process_analysis_chunked(INPUT_FILE, OUTPUT_FILE, CHUNK_SIZE)
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict

FILE = "../data/processed/ANALYSIS_ValeursFoncieres.csv"
CHUNK_SIZE = 100_000

# -----------------------------
# Aggregators
# -----------------------------
count_total = 0

sum_price = 0
sum_price_sq = 0

# Group stats
price_by_dep = defaultdict(float)
count_by_dep = defaultdict(int)

price_by_type = defaultdict(float)
count_by_type = defaultdict(int)

price_m2_sum = defaultdict(float)
price_m2_count = defaultdict(int)

# For histogram (approx)
price_sample = []
surface_sample = []
m2_sample = []

# -----------------------------
# Process chunks
# -----------------------------
for i, chunk in enumerate(pd.read_csv(FILE, chunksize=CHUNK_SIZE), start=1):

    # -------------------------
    # Feature engineering
    # -------------------------
    chunk["property_surface"] = (
        chunk["building_real_surface"].fillna(0) +
        chunk["land_surface"].fillna(0)
    )

    chunk["value_per_m2"] = chunk["property_value"] / chunk["property_surface"]
    chunk = chunk.replace([np.inf, -np.inf], np.nan)

    # -------------------------
    # Global stats
    # -------------------------
    valid_price = chunk["property_value"].dropna()

    count_total += len(valid_price)
    sum_price += valid_price.sum()
    sum_price_sq += (valid_price ** 2).sum()

    # -------------------------
    # Sample for histograms
    # -------------------------
    if len(price_sample) < 200_000:
        price_sample.extend(valid_price.values[:20_000])

    if len(surface_sample) < 200_000:
        surface_sample.extend(chunk["property_surface"].dropna().values[:20_000])

    if len(m2_sample) < 200_000:
        m2_sample.extend(chunk["value_per_m2"].dropna().values[:20_000])

    # -------------------------
    # Department aggregation
    # -------------------------
    grp_dep = chunk.groupby("dep_code")["property_value"].agg(["sum", "count"])
    for dep, row in grp_dep.iterrows():
        price_by_dep[dep] += row["sum"]
        count_by_dep[dep] += row["count"]

    # -------------------------
    # Property type aggregation
    # -------------------------
    grp_type = chunk.groupby("property_type")["property_value"].agg(["sum", "count"])
    for typ, row in grp_type.iterrows():
        price_by_type[typ] += row["sum"]
        count_by_type[typ] += row["count"]

    # -------------------------
    # €/m² aggregation
    # -------------------------
    valid = chunk.dropna(subset=["value_per_m2", "property_type"])
    grp_m2 = valid.groupby("property_type")["value_per_m2"].agg(["sum", "count"])

    for typ, row in grp_m2.iterrows():
        price_m2_sum[typ] += row["sum"]
        price_m2_count[typ] += row["count"]

    print(f"Chunk {i} processed")

# -----------------------------
# Final stats
# -----------------------------
mean_price = sum_price / count_total
std_price = np.sqrt((sum_price_sq / count_total) - mean_price**2)

print("\n===== GLOBAL STATS =====")
print(f"Mean price: {mean_price:,.0f} €")
print(f"Std price: {std_price:,.0f} €")

# -----------------------------
# Build series
# -----------------------------
df_dep = pd.Series({
    dep: price_by_dep[dep] / count_by_dep[dep]
    for dep in price_by_dep
}).sort_values()

df_type = pd.Series({
    typ: price_by_type[typ] / count_by_type[typ]
    for typ in price_by_type
}).sort_values()

df_m2 = pd.Series({
    typ: price_m2_sum[typ] / price_m2_count[typ]
    for typ in price_m2_sum
}).sort_values()

# -----------------------------
# PLOTS
# -----------------------------

# Price distribution
plt.figure(figsize=(10,5))
plt.hist(price_sample, bins=100)
plt.title("Property Value Distribution")
plt.show()

# Log price
plt.figure(figsize=(10,5))
plt.hist(np.log1p(price_sample), bins=100)
plt.title("Log Price Distribution")
plt.show()

# Surface distribution
plt.figure(figsize=(10,5))
plt.hist(surface_sample, bins=100)
plt.title("Surface Distribution")
plt.show()

# €/m² distribution
plt.figure(figsize=(10,5))
plt.hist(np.clip(m2_sample, 0, 20000), bins=100)
plt.title("€/m² Distribution")
plt.show()

# Price by department
df_dep.plot(kind="bar", figsize=(12,5))
plt.title("Average Price by Department")
plt.show()

# Price by property type
df_type.plot(kind="bar", figsize=(8,5))
plt.title("Average Price by Property Type")
plt.show()

# €/m² by type
df_m2.plot(kind="bar", figsize=(8,5))
plt.title("€/m² by Property Type")
plt.show()
````

### 🚀 1. Setup aggregators

````python
from collections import defaultdict
import pandas as pd
import numpy as np

FILE = "../data/processed/ANALYSIS_ValeursFoncieres.csv"
CHUNK_SIZE = 100_000

# Global aggregations
count_total = 0

# Numeric stats
sum_price = 0
sum_price_sq = 0

sum_surface = 0
sum_surface_sq = 0

# Group aggregations
price_by_dep = defaultdict(float)
count_by_dep = defaultdict(int)

price_by_type = defaultdict(float)
count_by_type = defaultdict(int)

price_per_m2_sum = defaultdict(float)
price_per_m2_count = defaultdict(int)
````

### 📦 2. Process chunks

````python
for chunk in pd.read_csv(FILE, chunksize=CHUNK_SIZE):

    # -------------------------
    # Basic cleanup
    # -------------------------
    chunk["property_surface"] = (
        chunk["building_real_surface"].fillna(0) +
        chunk["land_surface"].fillna(0)
    )

    chunk["value_per_m2"] = chunk["property_value"] / chunk["property_surface"]

    # Remove division issues
    chunk = chunk.replace([np.inf, -np.inf], np.nan)

    # -------------------------
    # Global stats
    # -------------------------
    valid_price = chunk["property_value"].dropna()
    valid_surface = chunk["property_surface"].dropna()

    count_total += len(valid_price)

    sum_price += valid_price.sum()
    sum_price_sq += (valid_price ** 2).sum()

    sum_surface += valid_surface.sum()
    sum_surface_sq += (valid_surface ** 2).sum()

    # -------------------------
    # Aggregation by department
    # -------------------------
    grp_dep = chunk.groupby("dep_code")["property_value"].agg(["sum", "count"])

    for dep, row in grp_dep.iterrows():
        price_by_dep[dep] += row["sum"]
        count_by_dep[dep] += row["count"]

    # -------------------------
    # Aggregation by property type
    # -------------------------
    grp_type = chunk.groupby("property_type")["property_value"].agg(["sum", "count"])

    for typ, row in grp_type.iterrows():
        price_by_type[typ] += row["sum"]
        count_by_type[typ] += row["count"]

    # -------------------------
    # €/m² aggregation
    # -------------------------
    valid = chunk.dropna(subset=["value_per_m2", "property_type"])

    grp_m2 = valid.groupby("property_type")["value_per_m2"].agg(["sum", "count"])

    for typ, row in grp_m2.iterrows():
        price_per_m2_sum[typ] += row["sum"]
        price_per_m2_count[typ] += row["count"]

    print(f"Processed {count_total:,} rows")
````

### 📊 3. Compute final stats

#### Global stats

````python
mean_price = sum_price / count_total
std_price = np.sqrt((sum_price_sq / count_total) - mean_price**2)

print("Mean price:", mean_price)
print("Std price:", std_price)
````

#### Department prices

````python
avg_price_dep = {
    dep: price_by_dep[dep] / count_by_dep[dep]
    for dep in price_by_dep
}

df_dep = pd.Series(avg_price_dep).sort_values()
````

#### Property type prices

````python
avg_price_type = {
    typ: price_by_type[typ] / count_by_type[typ]
    for typ in price_by_type
}

df_type = pd.Series(avg_price_type).sort_values()
````

#### €/m² by type

````python
avg_m2_type = {
    typ: price_per_m2_sum[typ] / price_per_m2_count[typ]
    for typ in price_per_m2_sum
}

df_m2 = pd.Series(avg_m2_type).sort_values()
````
