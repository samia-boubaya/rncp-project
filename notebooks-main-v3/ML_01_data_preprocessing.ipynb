## **Train-Test Split**

**Perform Train Test Split**
- split Valeurs Foncieres dataset into train and test sets (stratify by target variable to maintain class distribution)


**Define target (y)**
**Define features (X)**


````python
from duckdb import df
import pandas as pd
from sklearn.model_selection import train_test_split

# features: all columns except "property_value"
X = df.drop(columns=["property_value"])
y = df["property_value"] # target variable

X_train, X_test, y_train, y_test = train_test_split(
    X,                  # features
    y,                  # target variable
    test_size=0.2,      # 20% of the data for testing, 80% for training
    random_state=42,    # set random_state as 42 for reproducibility
    stratify=y)         # stratify the split to maintain class distribution in both sets

# save the training and testing sets to CSV files
X_train.to_csv("../data/processed/X_train.csv", index=False)
X_test.to_csv("../data/processed/X_test.csv", index=False)
y_train.to_csv("../data/processed/y_train.csv", index=False)
y_test.to_csv("../data/processed/y_test.csv", index=False)
````
