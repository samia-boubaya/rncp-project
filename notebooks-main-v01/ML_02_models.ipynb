## **STEP 3 — Reduce overfitting or underfitting**

**METHODS TO Check over/under-fitting**
1. **Train vs Test Performance**
- Compare metrics (R², MAE, RMSE) on train and test sets
    - Train ≫ Test → Overfitting
    - Both low → Underfitting

2. **Cross-Validation Scores**
- Compare train vs CV scores
    - Large gap → Overfitting
    - Both low → Underfitting
    - Stable & close → Good generalization

3. **Learning Curves**
- Plot performance vs training size
    - Gap between curves → Overfitting
    - Both curves low → Underfitting

4. **Residual Analysis**
- Plot residuals vs predictions
    - Patterns / structure → Underfitting
    - Large variance on test → Overfitting

### **Quick Rule of Thumb to check over/under-fitting**

| Situation           | TL;DR                     | Diagnosis    | Fix                           |
| ------------------- | ------------------------- | ------------ | ----------------------------- |
| Train >> Test       | Model memorizes data      | Overfitting  | Regularization, simpler model |
| Train ≈ Test (low)  | Model too simple          | Underfitting | More features, complex model  |
| Train ≈ Test (high) | Model generalizes well    | Good         | ✅                             |


---
### **Method 1: Compare Train vs Test Performance**    

**How to Interpret Train vs Test R² Results**

✅ Good Fit: Train ≈ Test performance
- **Good generalization**  Small difference → model generalizes well.

Example:
- Train R² = 0.82
- Test R² = 0.79

👉 Model generalizes well


🔴 Overfitting: Train much better than Test. 
- **Large gap → model memorized training data**
- Rule of Thumb → If(Train − Test > 5–7%) probably overfitting.

Example:
- Train R² = 0.95
- Test R² = 0.60

👉 Model memorizes training data but fails on new data



🔵 Underfitting: Both Train and Test are poor. 
- **Both low → model too simple**

Example:
- Train R² = 0.45
- Test R² = 0.40

👉 Model too simple, not capturing patterns
